```python
import h5py
```

```python
PATH_TO_AS_TRAIN_HDF5 = r'C:\Users\mp431591\Documents\work_code\cl_30\cl_data_as_train.hdf5'
PATH_TO_AS_EVAL_HDF5 = r'C:\Users\mp431591\Documents\work_code\cl_30\cl_data_as_eval.hdf5'
PATH_TO_FSD50K_TRAIN_HDF5 = r'C:\Users\mp431591\Documents\work_code\cl_30\cl_data_fsd50k_train.hdf5'
PATH_TO_FSD50K_EVAL_HDF5 = r'C:\Users\mp431591\Documents\work_code\cl_30\cl_data_fsd50k_eval.hdf5'
```

```python
# It's redundant to keep these datafiles as separate but the redundancy is
# by design to prevent losing or corrupting data.

as_hdf5_filename = "Audioset_data_cl.hdf5"

hdf5_as_train = h5py.File(PATH_TO_AS_TRAIN_HDF5, 'r')
hdf5_as_eval = h5py.File(PATH_TO_AS_EVAL_HDF5, 'r')
hdf5_fsd50k_train = h5py.File(PATH_TO_FSD50K_TRAIN_HDF5, 'r')
hdf5_fsd50k_eval = h5py.File(PATH_TO_FSD50K_EVAL_HDF5, 'r')

with h5py.File(as_hdf5_filename, 'a') as hdf5_as:
    grp_as_train = hdf5_as_train['audioset_strong_train_50']
    grp_as_eval = hdf5_as_eval['audioset_strong_eval_50']
    grp_fsd50k_train = hdf5_fsd50k_train['fsd50k_train_50']
    grp_fsd50k_eval = hdf5_fsd50k_eval['fsd50k_eval_50']

    grp_as_train.copy(source=grp_as_train, dest=hdf5_as)
    grp_as_eval.copy(source=grp_as_eval, dest=hdf5_as) 
    grp_fsd50k_train.copy(source=grp_fsd50k_train, dest=hdf5_as)
    grp_fsd50k_eval.copy(source=grp_fsd50k_eval, dest=hdf5_as)


hdf5_as_train.close()
hdf5_as_eval.close()
hdf5_fsd50k_train.close()
hdf5_fsd50k_eval.close()
```
